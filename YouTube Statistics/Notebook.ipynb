```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the csv file
df = pd.read_csv('Global YouTube Statistics.csv', encoding="latin1")
df.head()
```

```python
#gives basic info on dataset like mean, count, max, min, std
df.describe()
```

```python
df.info()
```

```python
#checking for NA values in your dataset
df.isna().sum()
```

```python
#checking for duplicates. there are no duplicates in our dataset
df.duplicated().sum()
```

```python
#seeing all the columns. 
df.columns
```

```python
avg_subscriber_for_each_category = df.groupby('category')['subscribers'].mean()
avg_subscriber_for_each_category
```

```python
#since the above numerical numbers are scientific 
# we can chagne that using pandas set_option method
pd.set_option('display.float_format','{:.2f}'.format)
```

```python
#now let's check that same data and see the change in numeric values
avg_subscriber_for_each_category
```

```python
plt.figure(facecolor='lightgreen')
plt.rcParams['axes.facecolor'] = 'black'
avg_subscriber_for_each_category.sort_values(ascending=False).plot(kind='barh', color='pink')
plt.title('Average Subscriber Distribution by Category')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

In the above graph we can easily see that 'Shows' has the most subscribers on YouTube followed by 'Trailers'.

```python
#getting number of channel types and then visually seeing the output
channel_type_count = df['channel_type'].value_counts()
plt.figure(facecolor='beige')
channel_type_count.plot(kind='barh', color='lightgreen')
plt.title('Channel Type Distribution')
plt.xlabel('Count')
plt.ylabel('Channel Type')
plt.show()
```

From the above graph we can surely say that most number of videos are of **Entertainment** type on yt.

```python
#let's create a scatter plot of subscribers and video views 
#showing how much views one gets with respect to no. of subscribers one has.
plt.figure(facecolor='beige')
plt.scatter(x = 'subscribers', y='video views', color='yellow', data = df)
plt.title("Scatter Plot: Subscribers vs Video Views")
plt.xlabel('Subscribers')
plt.ylabel('Video Views')
plt.show()
```

In the above figure we can see a **positive relationship** between subscribers and video views which means that more the subscribers, more the views. There are few exceptions obviously but those channels could be less active as well.

```python
#let's check total subscribers from each country
total_subscribers_by_country = df.groupby('Country')['subscribers'].sum().sort_values(ascending=False)
plt.figure(facecolor='beige', figsize=(10,6))
plt.xticks(rotation=90, ha='right')
plt.plot(total_subscribers_by_country,marker='.', color='orange')
plt.title('Total Subscribers from each country')
plt.xlabel('Country')
plt.ylabel('No. of Subscribers')
plt.show()
```

From the above graph it is clearly visible that ***USA*** has the most amount of subscribers on YouTube followed by India and Brazil. USA has a total of approx. **7.3 trillion subscirbers**. This is a total subscribers' number and we should interpret in this way that channel created by USA's citizens have most no. of total subscribers.

```python
#let's see the number of subscribers the channel has with respect to the year it was created in
df.groupby('created_year')['subscribers'].mean().sort_values(ascending=False)
```

This doesn't mean one cannot gain subs fast. It's the data that says total subs of the yt channels created in 1970 is the highest in number as compared to another years.

```python
plt.figure(facecolor='beige', figsize=(12,6))
df.groupby('Country')['Unemployment rate'].min().sort_values().plot(kind='bar')
plt.title('Unemployment rate of each country')
plt.ylabel('Count')
```

```python
#visually seeing the total number of uploads by each channel type
plt.figure(facecolor='lightblue')
df.groupby('channel_type')['uploads'].sum().sort_values(ascending=True).plot(kind='barh')
plt.title('Total Uploads by each type of channel')
plt.xlabel('Number of Uploads')
plt.ylabel('Channel type')
plt.show()
```

```python
#let's create a box plot for highest earning channel type
plt.figure(facecolor='beige')
plt.rcParams['axes.facecolor'] = 'lightgreen'
sns.boxenplot(x= 'channel_type', y='highest_monthly_earnings', data=df)
plt.title('Distribution of Highest Monthly Earning by channel type')
plt.xlabel('Channel tyoe')
plt.ylabel('Monthly Earnigns')
plt.xticks(rotation=90)
plt.show()
```

The most highest monthly earning is of **People** channel type

```python
#grouping by country and category and filtering for total subscirbers 
# in each category of each country
plt.figure(facecolor='black',figsize=(9,6))
plt.rcParams['axes.facecolor'] = 'lightblue'
df.groupby(['category', 'Country'])['subscribers'].sum().unstack().plot(kind='bar', stacked=True)
plt.title('Subscribers breakdown by category and country')
plt.xlabel('Country')
plt.ylabel('Total Subscribers')
plt.legend(title='Country', bbox_to_anchor=(1.05,1), loc = 'upper left')
plt.show()
```

```python
plt.figure(facecolor='beige')
plt.rcParams['axes.facecolor'] = 'black'
plt.scatter(df['Gross tertiary education enrollment (%)'], df['subscribers'], alpha=0.5)
plt.title('Scatter Plot: Gross Tertiary Education Percentage vs Subscribers')
plt.xlabel('Gross Tertiary Education Enrollment (%)')
plt.ylabel('No. of Subscribers')
plt.show()
```

```python
plt.figure(facecolor='beige')
df.groupby('channel_type')['Unemployment rate'].mean().sort_values().plot(kind='bar', color='lightgreen')
plt.title('AVG Unemployment rate distribution by channel type')
plt.xlabel('Channel Type')
plt.ylabel('Avg Unemployment rate')
plt.show()
```

On an average Animals type of channel shows most unemployment rate. One thing to note here is that unemployment rate was given for a country in our dataset. We can conclude that Animals type of channel are being created by the people of those countries where on an avg unemployment rate is high. 

```python
#grouping by created year and channel type
#filtering by subscribers
plt.figure(figsize=(10,6), facecolor='beige')
plt.rcParams['axes.facecolor'] = 'lightblue'
df.groupby(['created_year', 'channel_type'])['subscribers'].mean().unstack().tail().plot(kind='bar', width=1)
plt.title('Average Subscribers by created year and channel type')
plt.xlabel('Creted year')
plt.ylabel('Average Subscriber')
plt.legend(title = 'channel Type', bbox_to_anchor=(1.05,1), loc = 'upper left')
plt.xticks(rotation=45)
plt.show()
```

Using above chart we can interpret that in which year which channel type had most or least subscribers. So, in year 2021, Games channels were having most subscribers all over the world which makes sense since 2021 was the covid year and there was a boon in gaming sector. 

```python
#now let's see last 8-10 years trend of total subscribers number. did it rise or fall? 
df.groupby(['created_year', 'created_month'])['subscribers'].sum().unstack().tail().plot(kind='area', stacked=True)
plt.title('Subscribers Growth over time')
plt.xlabel('Year')
plt.ylabel('Total Subscribers')
plt.legend(title='Month', bbox_to_anchor = (1.05,1), loc = 'upper left')
plt.xticks(rotation=45)
plt.show()
```

We can clearly see that there is a huge fall in total subscribers in last decade.

```python
df.groupby(['created_year', 'Country'])['highest_yearly_earnings'].mean().unstack().tail().plot(kind='line', marker='.')
plt.title('Trends of Highest yearly earnings by Created Year and Country')
plt.xlabel('Created Year')
plt.ylabel('Average Highest Yearly Earnings')
plt.legend(title='Country', bbox_to_anchor = (1.05,1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

I think there is no proper trend and hence it is difficult to come to a conclusion from above line chart

```python
top_5_highest_monthly_earning_categories = df.groupby('category')['highest_monthly_earnings'].sum().sort_values(ascending=False).head()
top_5_highest_monthly_earning_categories
```
